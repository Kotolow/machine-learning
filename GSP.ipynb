## 3.1 GSP alg

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
data = ['TATTTATGGTATG','GTACTCTTAGTA','CCCGAGGTGAGCT','GTGTCACATATA','TCGTCTAAAAT']
def unique_val(data):
    temp=[]
    for i in range(len(data)):
        temp=temp+list(set(data[i]))
    unique_el=list(set(temp))
    unique_el.sort()
    return unique_el
I=unique_val(data)
MinsupValue = 4.0
MinsupRatio = MinsupValue/len(data)
print("Data: ", data)
print("Minsup value: ", MinsupValue)
```

```python
def compSup(data,I,k):
    Ck_dict={}
    for i in I:
        Ck_dict[i]=0
    for i in I:
        for j in data:
            counter=0
            m=0
            n=0
            while m<len(i) and n<len(j):
                if i[m]==j[n]:
                    counter+=1                   
                    m+=1
                    n+=1
                else:
                    n+=1
            if counter==len(i):
                Ck_dict[i]+=1
    print('Step: k=',k)
    print('After comp sup: ',Ck_dict)
    return Ck_dict
```

```python
def expand(Ck_dict,I,MinsupValue,F):
    Ck_list=[]
    reduced_Ck={}
    for i in range(len(I)):
        if(Ck_dict[I[i]]>=MinsupValue):
            Ck_list.append(I[i])
            print('Add to F: (',I[i],',',Ck_dict[I[i]],')')
            F[I[i]]=Ck_dict[I[i]]
            reduced_Ck[I[i]]=Ck_dict[I[i]]
    print('After reduce: ',reduced_Ck)
    Ck_expand=[]
    temp_list=[]
    for i in Ck_list:
        temp_list.append(i[len(i)-1])
    temp=set(temp_list)
    temp2=list(temp)
    for i in Ck_list:
        for j in temp2:
            Ck_expand.append(i+j)
    print('After expend: ',Ck_expand)    
    return Ck_expand
```

```python
def GSP(data,I,MinsupValue):
    print("Start GSP with minsup value:", MinsupValue)
    print("Unique el's: ", I)
    k=1
    F={}
    C1=compSup(data,I,k)
    Ck=expand(C1,I,MinsupValue,F)
    print()
    k=2
    while Ck!=[]:
        Sup=compSup(data,Ck,k)
        Ck=expand(Sup,Ck,MinsupValue,F)
        k+=1
        print()
    print('END GSP')
    return(F)
F=GSP(data,I,MinsupValue)
print()
print('Result table:')
F_list=list(F.keys())
count = len(F_list)-15
for i in range(len(F_list)-15,len(F_list)):
    print(count, F[F_list[i]], end= " ")
    count = count + 1
    for l in F_list[i]:
        print(l, end="")
    print(" ", len(F_list[i]))
```
